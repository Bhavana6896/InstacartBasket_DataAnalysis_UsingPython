# 7. Frequency Flags

```python
ords_prods_merge_new['ord_regularity_median'] = ords_prods_merge_new.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
ords_prods_merge_new[['user_id', 'prices', 'days_since_last_order', 'ord_regularity_median']].head(25)
```

### Types 

```python
ords_prods_merge_new.loc[ords_prods_merge_new['ord_regularity_median'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge_new.loc[(ords_prods_merge_new['ord_regularity_median'] > 10) & (ords_prods_merge_new['ord_regularity_median'] <= 20), 'frequency_flag']  = 'Regular customer'
```

```python
ords_prods_merge_new.loc[ords_prods_merge_new['ord_regularity_median'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

### Frequency

```python
ords_prods_merge_new['frequency_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge_new[['user_id', 'days_since_last_order', 'ord_regularity_median', 'frequency_flag']].head(25)
```

```python
ords_prods_merge_new[['user_id', 'days_since_last_order', 'ord_regularity_median', 'frequency_flag']].tail(25)
```

### Answer: There are 21559853 frequent customers who can be targeted on the Instacart App to send them notifications for ads and offers.
