# 5. Merging Data

```python
print(df_orders_products_combined.columns)
print(df_products.columns)
```

```python
print(df_orders_products_combined['product_id'].dtype)
print(df_products['product_id'].dtype)
```

```python
df_orders_products_combined = df_orders_products_combined.drop(columns=['_merge'], errors='ignore')
df_products = df_products.drop(columns=['_merge'], errors='ignore')
```

```python
df_ords_prods_merge = df_orders_products_combined.merge(df_products, on='product_id', how='inner', indicator=True)
```
