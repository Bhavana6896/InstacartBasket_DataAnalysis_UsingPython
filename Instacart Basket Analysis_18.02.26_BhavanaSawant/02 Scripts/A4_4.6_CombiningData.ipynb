# Concatenating Data

```python
df_concat
```

####  This is the axis argument. Remember that the default setting for pd.concat() is to stack any new rows after the existing ones (creating a long-format dataframe). The argument for this is axis = 0, but since this is the default, you don’t need to specify it. If, however, you wanted to place the two dataframes side by side (creating a wide-format dataframe), you’d need to use a different argument: axis = 1.

```python
df_concat = pd.concat(frames, axis = 1)
```

```python
df_concat 
```

##### the pd.concat() function:

Is suitable for rows or columns of the same length.
Will place dataframes on top of each other by default (axis = 0).
Requires a list as its main argument (this is why the frames list was created first in the example above).

# Merging Data

##### The best use cases for the df.merge() function are those where the dataframes you want to combine don’t match in shape—different from concatenate. In these cases, you’ll need a key or some kind of common identifier column that brings the two (or more) data sets together.

```python
data3 = {'customer_id':['6732', '767', '890', '635'],
    'month':['Jan-20', 'Jan-20', 'Jan-20', 'Jan-20'],
    'days_purchased_on':[0, 10, 4, 1]}
```

```python
df_2 = pd.DataFrame(data3,index=[0, 1, 2, 3])
```

```python
df_merged = df.merge(df_2, on = 'customer_id')
```

```python
df_merged 
```

# In case you want two keys, you can use:

df_merged = df.merge(df_2, on 0 = ['customer_id', 'month'])

Note here the syntax for using multiple columns as keys during a merge. The columns are included in the form of a list inside square brackets within the on argument. This is necessary when including multiple parameters within a single argument. Multiple arguments within a single function are separated by commas, but multiple parameters within a single argument need to be set off in brackets. If you don’t include the brackets and simply offset your parameters with commas (df.merge(df_2, on = 'customer_id', 'month')), pandas will think these are additional arguments, resulting in errors.

```python
df_merged2 = df.merge(df_2, on = ['customer_id', 'month'])
```

```python
df_merged2 
```

###### A quick and easy way to check for a full match is via the indicator = True argument.

```python
df_merged = df.merge(df_2, on = ['customer_id', 'month'], indicator = True)
```

```python
df_merged
```

```python
df_merged['_merge'].value_counts()
```
