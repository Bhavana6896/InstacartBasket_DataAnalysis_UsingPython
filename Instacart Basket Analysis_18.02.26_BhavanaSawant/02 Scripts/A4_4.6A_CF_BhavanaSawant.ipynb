# 2. Exporting in other formats

## A pickle, or “.pkl,” is a pandas format used to store data on your computer. Pickles can only be opened using Python.

```python
# Export data to pkl

df_merged_large.to_pickle(os.path.join(path, '02 Data','Prepared Data', 'orders_products_combined.pkl'))
```

```python
# Export data to csv

df_merged_large.to_csv(os.path.join(path, '02 Data','Prepared Data', 'orders_products_combined.csv'))
```

### The only difference is the use of function: in the first command, the function to_csv() is used, while in the second command, the function to_pickle() is used. Pay attention to the file extension and ensure it matches the function you use!

Importing pickle files also follows a similar syntax to its “.csv” counterpart: the only difference comes in the function (read_pickle()) and the lack of an index_col, since pickle-format files include this information already.
